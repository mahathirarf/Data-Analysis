```python
for i in data.columns:
  if data[i].dtypes == 'O':
    print('---------------------')
    print(data[i].value_counts())
    print('---------------------')
```

```python
categorical = []

for i in data.columns:
  if data[i].dtypes == 'O':
    categorical.append(i)
```

```python
numerical = []

for i in data.columns:
  if data[i].dtypes != 'O':
    numerical.append(i)
```

```python
for i in numerical:
  sns.displot(data[i])
  plt.show()
```

```python
for i in categorical:
  sns.countplot(data=data, x=i)
  plt.show()
```

```python
round(data[numerical].corr(), 3)
```

```python
contingency_table = pd.crosstab(data['Country'], data['Diet Type'])
```

```python
from itertools import permutations, combinations
pairs = list(combinations(categorical, 2))
print(pairs)
```

```python
import scipy.stats as stats
```

```python
for i in pairs:
  contingency_table = pd.crosstab(data[i[0]], data[i[1]])
  chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
  n = contingency_table.sum().sum()
  cramers_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))
  print('For ' + i[0] + ' and ' + i[1] + ' Cramers V is ' + str(cramers_v))
```
